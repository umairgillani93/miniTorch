```python
import os 
import sys 
import torch
import random
from torchviz import make_dot
from torch.optim import SGD
random.seed(42)

x = torch.randn((5,5), requires_grad = True)
y = 2 * x + 5
loss = y.sum()
loss.backward()

# Tensor Struct { shape, stride, ndim, Op, parents, grad} # theta_new = theta_old - lr * partial_deri(cost)

make_dot(loss, params = {'x': x})
```
